```python
### I guess what we can do is a really really short version?
### what if we have just one image?

def make_short(gt, count):

    gt_short = {}
    gt_short['categories'] = gt['categories']
    gt_short['images'] = []
    gt_short['annotations'] = []

    for tmp in gt['images']:
        if tmp['id'] < count:
            gt_short['images'].append(tmp)
    
    
    for tmp in gt['annotations']:
        if tmp['image_id'] < count:
            gt_short['annotations'].append(tmp)
        else:
            break
    return gt_short
```
